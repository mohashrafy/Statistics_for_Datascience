# Quantitative Data :  Histograms

### what quantitative variables are

They are numerical values, known as quantities, on which we can execute mathematical operations.

Height, weight, wealth, test results, shoe size, or the number of heads after ten coin flips are some examples.

Also, keep in mind that we have both **discrete** and continuous **quantitative** variables.

So a discrete random variable might be the number of heads after 10 coin flips, where it must take an integer value.

Height and weight, on the other hand, can take any value, making them a continuous variable.

So you can be 5.6 , 5.7, 5.8  or any figure in between. 

### why do we use histograms? 

They provide an excellent *first* analysis of our data.

And we can get a brief overview of how our data looks and what we might want to do with it. 

```python
# Importing necessary mdoules 
import pandas as pd
import seaborn as sns
# Create Data Frame from data
male_heights_inch = [73.84701702, 68.78190405, 74.11010539, 71.7309784, 69.88179586,
          67.25301569, 68.78508125, 68.34851551, 67.01894966, 63.45649398,
          71.19538228, 71.64080512, 64.76632913, 69.2830701, 69.24373223,
          67.6456197,  72.41831663, 63.97432572, 69.6400599, 67.93600485,
          67.91505019, 69.43943987, 66.09017738, 66.14913196, 66.99094408]
df = pd.DataFrame(data={"heights": male_heights_inch})
df.heights = df.heights * 2.54
df
```

```python
sns.histplot(data=df["heights"], bins=10, kde=True)
```

### 4 main aspects of Histogram
* **Shape**: The shape of the histogram refers to its overall look.

  **symmetric, bell-shaped, skewed to the left or right** 
* **Center**: mean or median
* **Spread**: how far our data reaches, **range**, **interquartile range(IQR)**, **standard deviation** or the **variance**.
* **Outliers**: data points that are separated from the majority of the data 
---

## 1.

1. **Shape** : a distribution that is approximately **bell-shaped**. This is **unimodal**, which means it only has one mode, and we only observe one peak in our shape. 
2. **Center**:  because we have a bell-shaped and it's also symmetric. The **Median** would just be right down the middle which we'll call at about -13.because this is symmetric, we would also have the **mean** being -13 as well, approximately. 
3. **Spread**:  $Range = Max(x_i) - Min(x_i) = 7 - (-33) = 40 $
4. **Outlier**: Because all of the data is rather concise and consistent, we may conclude that there are **no obvious outliers**. 

**Putting it All together**:  the distribution of Ideology in American States is roughly bell shaped with a center of about -13 in Ideology Score. A range of 40 score, (-33 to 7), and no apparent outliers.

## 2.

Let's make a polt from Kaggle for the SF Salaries. 
[source](https://www.kaggle.com/kaggle/sf-salaries?select=Salaries.csv)

```python
sal = pd.read_csv("data/Salaries.csv")
```

```python
sns.histplot(data=sal["TotalPay"], bins="auto", kde=True, color="teal")
```

```python
sal["TotalPay"].describe()
```

1. **Shape** :  **right skewed, bimodal** distribution . 
2. **Center**: median would be about 80,000 dollar. Because it is right skewed,that means the mean is going to be pulled to the right.let's call it 85,000 dollar is our mean for this data

3. **Spread**:  Range is simply going to be 600,000 dollar
4. **Outlier**: maybe pass 200,000 dollar where we consider it an outlier. 

**Putting it All together**:  Salary distribution in San Francisco is bimodal and tilted to the right. The data is centered around 80,000 dollar , with the majority of the data falling between 40,000 dollar and 120,000 dollar.
A price range of around 600,000 dollar, with outliers on the higher end.
