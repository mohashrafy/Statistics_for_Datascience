# Quantitative Data: Numerical Summaries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data and store it in the Data Frame
scores = pd.read_csv("../data/scores.csv")
# Let's look at the data
print(scores.shape)
scores.head()
```

```python
# Create Histogram of math scores
sns.histplot(data=scores['Average Score (SAT Math)'], kde=True)
# Draw Mean with Red line
plt.axvline(x=scores['Average Score (SAT Math)'].mean(), color='red')
# Draw Mean with Yellow line
plt.axvline(x=scores['Average Score (SAT Math)'].median(), color='yellow')
```

```python
# Create Numerical Summaries for math score
scores['Average Score (SAT Math)'].describe().to_frame()
```
