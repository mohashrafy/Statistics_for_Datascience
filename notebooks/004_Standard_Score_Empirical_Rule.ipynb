## **Rule of 68-95-99.7**
# 1. One Standard Deviation

```python
# Create Histogram of Sleep Hours
ax = sns.histplot(data=sleep['Hours'], bins=9, kde=True, color="blue")
xs, ys = ax.lines[0].get_data()
# Draw Mean with Red line
mean = sleep['Hours'].mean()
ax.axvline(x=mean, color='red')
# Calculate Standard Deviation
sdev = sleep['Hours'].std()
# Calculate one Std from the mean
left_1sd = mean - sdev
right_1sd = mean + sdev
print(f"One Std to the left = {left_1sd:.2f}")
print(f"One Std to the right = {right_1sd:.2f}")
# Draw one Standard Deviation right and left with yellow line
ax.axvline(x=left_1sd, color='yellow')
ax.axvline(x=right_1sd, color='yellow')
# Fill between +1 and -1 Standard Deviation from the mean with red
ax.fill_between(xs,
                ys,
                where=(xs > left_1sd) & (xs < right_1sd),
                interpolate=True,
                facecolor='red',
                alpha=1)
# Fill outer part of distribution with blue
ax.fill_between(xs,
                ys,
                where=(xs < left_1sd) | (xs > right_1sd),
                interpolate=True,
                facecolor='blue',
                alpha=0.9)
```

### The red area contains 68% of the data. 

$-1 Std$ **<------------------------(68%)------------------------>** $+1 Std$

$5.24$ Hour **<--------(68% People sleep between)-------->** $8.07$ Hour

We would know that for bell-shaped or normal distributions 68 percent of the values are expected to be
within one standard deviation away from the mean. 

---

# 2. Two Standard Deviation

```python
# Create Histogram of Sleep Hours
ax = sns.histplot(data=sleep['Hours'], bins=9, kde=True, color="green")
xs, ys = ax.lines[0].get_data()
# Draw Mean with Red line
mean = sleep['Hours'].mean()
ax.axvline(x=mean, color='red')
# Calculate Standard Deviation
sdev = sleep['Hours'].std()
# Calculate two Std from the mean
left_2sd = mean - 2 * sdev
right_2sd = mean + 2 * sdev
print(f"Two Std to the left = {left_2sd:.2f}")
print(f"Two Std to the right = {right_2sd:.2f}")
# Draw two Standard Deviation right and left with yellow line
ax.axvline(x=left_2sd, color='yellow')
ax.axvline(x=right_2sd, color='yellow')
# Fill between +2 and -2 Standard Deviation from the mean with red
ax.fill_between(xs,
                ys,
                where=(xs > left_2sd) & (xs < right_2sd),
                interpolate=True,
                facecolor='red',
                alpha=1)
# Fill outer part of distribution with green
ax.fill_between(xs,
                ys,
                where=(xs < left_2sd) | (xs > right_2sd),
                interpolate=True,
                facecolor='green',
                alpha=0.9)
```

### The red area contains 95% of the data. 

$-2 Std$ **<------------------------(95%)------------------------>** $+2 Std$

$3.82$ Hour **<--------(95% People sleep between)-------->** $9.49$ Hour

We would know that for bell-shaped or normal distributions 95 percent of the values are expected to be
within two standard deviation away from the mean.

# 3. Three Standard Deviation

```python
# Create Histogram of Sleep Hours
ax = sns.histplot(data=sleep['Hours'], bins=9, kde=True, color="purple")
xs, ys = ax.lines[0].get_data()
# Draw Mean with Red line
mean = sleep['Hours'].mean()
ax.axvline(x=mean, color='red')
# Calculate Standard Deviation
sdev = sleep['Hours'].std()
# Calculate three Std from the mean
left_3sd = mean - 3 * sdev
right_3sd = mean + 3 * sdev
print(f"Three Std to the left = {left_3sd:.2f}")
print(f"Three Std to the right = {right_3sd:.2f}")
# Draw three Standard Deviation right and left with yellow line
ax.axvline(x=left_3sd, color='yellow')
ax.axvline(x=right_3sd, color='yellow')
# Fill between +3 and -3 Standard Deviation from the mean with red
ax.fill_between(xs,
                ys,
                where=(xs > left_3sd) & (xs < right_3sd),
                interpolate=True,
                facecolor='red',
                alpha=1)
# Fill outer part of distribution with green
ax.fill_between(xs,
                ys,
                where=(xs < left_3sd) | (xs > right_3sd),
                interpolate=True,
                facecolor='green',
                alpha=0.9)
```

### The red area contains 99.7% of the data. 

$-3 Std$ **<------------------------(99.7%)------------------------>** $+3 Std$

$2.40$ Hour **<--------(99.7% People sleep between)-------->** $10.91$ Hour

We would know that for bell-shaped or normal distributions 99.7 percent of the values are expected to be
within two standard deviation away from the mean.

## Standard Score:
In statistics, the standard score is the number of standard deviations by which the value of a raw score (i.e., an observed value or data point) is above or below the mean value of what is being observed or measured.Standard scores are most commonly called **z-scores**.

Formula:

## $Z = \frac{x - \mu}{\sigma}$

Z = standard score

x = observed value

$\mu$ =  mean of the sample

$\sigma$ = standard deviation of the sample
